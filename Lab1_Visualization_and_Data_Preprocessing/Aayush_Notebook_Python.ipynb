```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")

df.info()
df.describe(include="all")
```

```python
numeric_cols = df.select_dtypes(include=np.number)

numeric_summary = numeric_cols.agg(
    ['mean', 'median', 'min', 'max', 'std', 'var']
).T

numeric_summary
```

```python
range_summary = numeric_cols.max() - numeric_cols.min()
range_summary
```

```python
categorical_cols = df.select_dtypes(include='object')

categorical_modes = categorical_cols.mode().iloc[0]
categorical_modes
```

```python
placement_counts = df['Placement_Status'].value_counts()
placement_percentages = df['Placement_Status'].value_counts(normalize=True) * 100

placement_counts, placement_percentages
```

```python
missing_summary = df.isna().sum()
missing_summary
```

```python
df['CGPA'].mean(), df['CGPA'].median()
```
