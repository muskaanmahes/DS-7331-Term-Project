# Preprocessing Summary

```python
import pandas as pd
import numpy as np

def summarize_dataset(df, top_n_categories=5):
    """
    Generates a preprocessing summary for a pandas DataFrame.

    Parameters
    ----------
    df : pandas.DataFrame
        Dataset to summarize
    top_n_categories : int
        Number of most frequent categories to display

    Returns
    -------
    summary_df : pandas.DataFrame
        Column-level summary
    """

    print("="*60)
    print("DATASET OVERVIEW")
    print("="*60)
    print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
    print(f"Duplicate Rows: {df.duplicated().sum()}")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    print("\n")

    summary = pd.DataFrame({
        "Data Type": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Missing %": (df.isnull().mean() * 100).round(2),
        "Unique Values": df.nunique()
    })

    # Numeric statistics
    numeric_cols = df.select_dtypes(include=np.number).columns
    if len(numeric_cols) > 0:
        stats = df[numeric_cols].describe().T[
            ["mean", "std", "min", "25%", "50%", "75%", "max"]
        ]
        summary = summary.join(stats, how="left")

    print("="*60)
    print("COLUMN SUMMARY")
    print("="*60)
    display(summary)

    # Categorical overview
    cat_cols = df.select_dtypes(include=["object", "category"]).columns

    if len(cat_cols) > 0:
        print("\n" + "="*60)
        print("CATEGORICAL COLUMN PREVIEW")
        print("="*60)

        for col in cat_cols:
            print(f"\nColumn: {col}")
            print(df[col].value_counts().head(top_n_categories))

    return summary
```

## Job Placement Dataset

```python
df = pd.read_csv("../Student Placement Dataset/full_dataset.csv")
summary = summarize_dataset(df)
```
